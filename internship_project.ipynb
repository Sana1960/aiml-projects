```python
#DIAMOND DATASET
# Step 1: Import necessary libraries
import seaborn as sns         # for loading the dataset
import pandas as pd           # for data handling
import matplotlib.pyplot as plt  # for plotting

# Step 2: Load the diamond dataset
df = sns.load_dataset('diamonds')

# Step 3: Display the first 5 rows
print("First 5 rows of the dataset:")
print(df.head())

# Step 4: Dataset info
print("\nDataset Info:")
print(df.info())

# Step 5: Summary statistics
print("\nSummary Statistics:")
print(df.describe())

# Step 6: Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())

# Step 7: Count of different cut types
print("\nCount of each cut type:")
print(df['cut'].value_counts())

# Step 8: Visualization - Price distribution
plt.figure(figsize=(8,5))
sns.histplot(df['price'], kde=True, color='skyblue')
plt.title("Distribution of Diamond Prices")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()

# Step 9: Boxplot - Price vs Cut
plt.figure(figsize=(8,5))
sns.boxplot(x='cut', y='price', data=df)
plt.title("Boxplot of Price by Cut")
plt.show()

# Step 10: Correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
